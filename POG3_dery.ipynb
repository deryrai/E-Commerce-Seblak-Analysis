```python
outliers = df[(df['harga_produk'] < lower_bound) | (df['harga_produk'] > upper_bound)]

no_outlier = df[(df['harga_produk'] >= lower_bound) & (df['harga_produk'] <= upper_bound)]
```

```python

print("Jumlah outlier:", len(outliers))
print("Jumlah non outlier:", len(no_outlier))
```

## Confidence Interval

```python
std = df['harga_produk'].std()     # Standard deviation of harga_produk
N = len(df)                    # Length of the data
low, up = stats.norm.interval(    # Will have lower & upper limit
    0.95,                         # Confidence level in 95%
    loc=df['harga_produk'].mean(), # Average of harga_produk that store in warehouse
    scale=std/np.sqrt(N)          # The width of dist. from average
    )         
print('Lower Limit:',low)   # Lower limit
print('Upper Limit:',up)    # Upper limit
```

# Uji hipotesis

## Uji hipotesis apakah bahan baku diluar jabodetabek mempengerahui harga jual seblak

H0 = Harga bahan baku diluar jabodetabek mempengaruhi harga jual seblak

H1 = Harga bahan baku diluar jabodetabek tidak mempengaruhi harga jual seblak

```python
t_stat, p_val = stats.ttest_ind(A['harga_produk'],B['harga_produk'])
print('T-Statistic:',t_stat)
print('P-value:',p_val) 
```

dikarnakan P-value lebih dari 0.05, maka H0 ditolak, yang artinya harga bahan baku tidak mempengaruhi harga jual seblak

```python
t_stat, p_val = stats.ttest_ind(df_joined['harga_produk'],df_joined['banyaknya_terjual'])
print('T-Statistic:',t_stat)
print('P-value:',p_val) 
```

Berdasarkan P value yang didapat lebih dari 0.05, maka H0 ditolak dan H1 diterima yaitu Harga produk tidak mempengaruhi nilai penjualan
